## Sending a Basic Prompt to the Model

Here's the simple chat flow to Gemma once again.

```python
from ollama import chat
from ollama import ChatResponse

model = 'gemma4:latest'

def model_call(model_prompt):
    response: ChatResponse = chat(model=model, messages=[
      {
        'role': 'user',
        'content': model_prompt,
      },
    ])
    return response['message']['content']

user_prompt = "Say hello to the class"
model_call(user_prompt)
```

## Adding an Image
Gemma 4 has been trained with multimodality, where images are converted into embedding vectors the model can operate on. As a user adding an image is quite straightforward.

```python
image_path = "img/ducks.jpg"  

response = chat(
        model="gemma4:latest",  
        messages=[
            {
                'role': 'user',
                'content': 'What is this?',
                'images': [image_path]
            }
        ]
    )

response["message"]["content"]
```

## OCR Use Cases

```python
image_path = "img/Receipt.jpg"

def image_chat(prompt, img_path):
    response = chat(
        model="gemma4:latest",
        messages=[
            {
                'role': 'user',
                'content': prompt,
                'images': [img_path]
            }
        ]
    )
    return response["message"]["content"]

response = image_chat("What's the text in this image?", image_path)
```
